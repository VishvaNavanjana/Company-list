```python
#file address 
file = "list.txt"
```

```python

#store the data in a list
company_list = []

#read the file line by line
with open(file) as fp:
    Lines = fp.readlines()
    for line in Lines: 
        company_list.append(line.strip())

print("Number of companies in the file : {}".format(len(company_list)))

#convert list to a set to get a unique company set
company_set = set(company_list)

print("Number of unique companies : {} \n".format(len(company_set)))

for company in company_set:
    print(company)

#write the python set to a csv file
with open("Unique_list.csv", "w") as fp:
        fp.write("Number of unique companies : {}\n".format(len(company_set)))
        for item in company_set:
                fp.write("%s\n" % item)
```
